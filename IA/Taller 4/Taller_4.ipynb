### a. Implementar una heurística consistente en la suma de las distancias de Manhattan (caminos por filas y columnas) de cada pieza del puzzle a su posición correcta en el estado solución:

Estado inicial a utilizar del puzzle para los metodos dados


```python
puzzle_array=np.array([[5,8,0],[2,1,4],[7,3,6]])
puzzleB=Puzzle8_State(None,0,puzzle_array)
print (puzzleB.getPosition())
```

"Primero el mejor"-Heurística Manhatan

```python
%%time
t_inicial = datetime.datetime.now()
puzzle_man=Puzzle8_State(None,0,puzzleB.getPosition())
puzzle_man.heuristic_val=heuristic_manh(puzzle_man)
solution_man=Puzzle8_bestfirst_inf(puzzle_man)
t_final = datetime.datetime.now() 
t_first_manh = t_final-t_inicial
t_first_manh=float(str(t_first_manh.seconds)+'.'+str(t_first_manh.microseconds))
print('Tiempo transcurrido %s Segundos '%(t_first_manh))
```

### b. Implementar la solución al puzzle de 8 casillas mediante el método A* usando la heurística propuesta.

```python
%%time
t_inicial = datetime.datetime.now()
puzzle_AA_=Puzzle8_State(None,0,puzzleB.getPosition())
puzzle_AA_.heuristic_val=heuristic_manh(puzzle_AA_)
solution_aa=Puzzle8_AA_(puzzle_AA_)
t_final = datetime.datetime.now() 
t_aa = t_final-t_inicial
t_aa=float(str(t_aa.seconds)+'.'+str(t_aa.microseconds))
print('Tiempo transcurrido %s Segundos '%(t_aa))
```

## Generar tabla objetivo (KPI)

```python
primer_mes=pd.DataFrame(df2.groupby(["ID"])["MONTHS_BALANCE"].agg(min))
primer_mes=primer_mes.rename(columns={'MONTHS_BALANCE':'Mes_Apertura'})

datos=pd.merge(df,primer_mes,on="ID")

df2['Seleccionador'] = None
df2['Seleccionador'][df2['STATUS'] =='2']='Yes' 
df2['Seleccionador'][df2['STATUS'] =='3']='Yes' 
df2['Seleccionador'][df2['STATUS'] =='4']='Yes' 
df2['Seleccionador'][df2['STATUS'] =='5']='Yes'

cpunt=df2.groupby('ID').count()

cpunt['Seleccionador'][cpunt['Seleccionador'] > 0]='Si' 
cpunt['Seleccionador'][cpunt['Seleccionador'] == 0]='No'

cpunt = cpunt[['Seleccionador']]

datos=pd.merge(datos,cpunt,how='inner',on='ID')
datos['Objetivo']=datos['Seleccionador']

datos.loc[datos['Objetivo']=='Si','Objetivo']=1
datos.loc[datos['Objetivo']=='No','Objetivo']=0
```
